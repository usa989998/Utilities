### Get data at a custom frequency


you can use the pandas resample method to convert the stock market data to the frequency of your choice. The implementation of these is shown below where a 1-minute frequency data is converted to 10-minute frequency data.


The first step is to define the dictionary with the conversion logic. For example, to get the open value the first value will be used, to get the high value the maximum value will be used and so on.

```python
ohlcv_dict = {
 '1. open': 'first',
 '2. high': 'max',
 '3. low': 'min',
 '4. close': 'last',
 '5. volume': 'sum'
}
```

Convert the index to datetime timestamp as by default string is returned. Then call the resample method with the frequency such as

10T for 10 minutes

D for 1 day and

M for 1 month

```python
intraday_data.index = pd.to_datetime(intraday_data.index)
```

```python
intraday_data
```

```python
intraday_data_10 = intraday_data.resample('10T').agg(ohlcv_dict)
intraday_data_10.head()
```

```python
intraday_data_10.tail()
```

### Import Data with yFinance 




The yfinance module has the download method which can be used to download the stock market data. It takes the following parameters:

####  ticker: The name of the tickers you want the data for. If you want data for multiple tickers then separate them by space

#### period: The number of days/month of data required. The valid frequencies are 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max

#### interval: The frequency of data. The valid intervals are 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo

```python
import yfinance as yf 
data = yf.download(tickers="MSFT", period="5d", interval="1m")
data.head()
```

### NSEpy
The nsepy package is used to get the stock market data for the futures and options for Indian stocks and indices.


```python
!pip install nsepy
```

```python
from datetime import date
from nsepy import get_history

# Stock options (for index options, set index = True)
stock_fut = get_history(symbol="HDFC", start=date(2019, 1, 15), end=date(2019, 2, 1), futures=True, expiry_date=date(2019, 2, 28))
stock_fut.head()
```

```python
import matplotlib.pyplot as plt
stock_fut["Close"].plot(figsize=(10, 5))

# Define the label for the title of the figure
plt.title("Close Price", fontsize=16)

# Define the labels for x-axis and y-axis
plt.ylabel('Price', fontsize=14)
plt.xlabel('Date', fontsize=14)

# Plot the grid lines
plt.grid(which="major", color='k', linestyle='-.', linewidth=0.5)

plt.show()
```
