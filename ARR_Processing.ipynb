```python
dw_mapping={
    0: 'Monday', 
    1: 'Tuesday', 
    2: 'Wednesday', 
    3: 'Thursday', 
    4: 'Friday',
    5: 'Saturday', 
    6: 'Sunday'
} 
```

```python
df['Value Date'] = pd.to_datetime(df['Value Date'], errors='coerce', dayfirst=True, format="%d/%m/%Y")
df
```

### Adding a Weekday By Name for Value Date

### Adding Weight Column as the difference between 2 consecutive dates

```python
df['Value Date - Weekday'] = pd.to_datetime(df['Value Date']).dt.weekday.map(dw_mapping)
df
```

```python
daycount_365, daycount_360, rate_divisor = 365.00, 360.00, 100.00

df['weight'] = df['Value Date'].shift(-1).dt.dayofyear - df['Value Date'].dt.dayofyear
```

```python
df['Rolling_Sum'] = df.weight.rolling(window=300, min_periods=1).sum()
```

```python
df["cum_Int_wt_A"] = df.weight.rolling(window=20).sum()
df["cum_Obs_wt_A"] = df.weight.shift(5).rolling(window=20).sum()

df["cum_Int_wt_B"] = df.weight.rolling(window=21).sum()
df["cum_Obs_wt_B"] = df.weight.shift(5).rolling(window=21).sum()

df["cum_Int_wt_C"] = df.weight.rolling(window=22).sum()
df["cum_Obs_wt_C"] = df.weight.shift(5).rolling(window=22).sum()
```
