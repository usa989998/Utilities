```python
daycount, rate_divisor = 365.00, 100.00
df_test["cum_Int_wt_3"] = df_test.Int_wt.rolling(window=3, min_periods=1).sum()
df_test["cum_Obs_wt_3"] = df_test.Obs_wt.rolling(window=3, min_periods=1).sum()
```

```python
df_test["daily_EAR"] = (df_test["rate"]/daycount)/rate_divisor
df_test["Int_wt_EAR"] = df_test["daily_EAR"] * df_test["Int_wt"]
df_test["Obs_wt_EAR"] = df_test["daily_EAR"] * df_test["Obs_wt"]
```

```python
df_test["Int_wt_Cap_Mult"] = df_test["Int_wt_EAR"] + 1.000000
df_test["Obs_wt_Cap_Mult"] = df_test["Obs_wt_EAR"] + 1.000000

df_test["Int_wt_Cap_Mult_Running_3"] = df_test['Int_wt_Cap_Mult'].rolling(3).apply(lambda x:x.prod())
df_test["Obs_wt_Cap_Mult_Running_3"] = df_test['Obs_wt_Cap_Mult'].rolling(3).apply(lambda x:x.prod())
```

```python
df_test
```

```python
df_test["Int_Compound_Rate"] = (df_test["Int_wt_Cap_Mult_Running_3"] - 1.000000) * (daycount/df_test["cum_Int_wt_3"])
df_test["Obs_Compound_Rate"] = (df_test["Obs_wt_Cap_Mult_Running_3"] - 1.000000) * (daycount/df_test["cum_Obs_wt_3"])
```

```python
df_test["Delta_Rate_Obs_Int"] = df_test["Obs_Compound_Rate"] - df_test["Int_Compound_Rate"] 
```
